```python
class Hero:
    '''
    A hero who should not be fed poison (like anyone else)
    but who resurrects when fed ham (unlike anyone else)
    '''
    def __init__(self, name):
        self.name= name
        self.health= 100
        print "New hero created named " + name
        
    def eat(self, food):
        if (food=='poison'):
            self.health -= 100
        elif (food=='ham'):
            self.health += 20
        if (self.health <= 0):
            print self.name + " is dead :("
            self.health= 0
            
bob= Hero('Bob')
```

```python
bob.eat('poison')
```

```python
bob.eat('ham')
print bob.health
```
