```python
def grab_state_hpi_data():

    main_df= pd.DataFrame()

    for abr in state_abbr:
        query= "FMAC/HPI_" + str(abr)
        print query
        df= Quandl.get(query, authtoken= api_key)
        # for each record, check its pct diff from the origin (1975 data)
        df[abr]= (df[abr]-df[abr][0]) * 100.0 / df[abr][0]

        if main_df.empty:
            main_df= df
        else:
            main_df= main_df.join(df)

        main_df.head()

        main_df.to_pickle('states_hpi.pickle')
```

```python
grab_state_hpi_data()
hpi_data= pd.read_pickle('states_hpi.pickle')
```

```python
def hpi_benchmark():
    df= Quandl.get("FMAC/HPI_USA", authtoken= api_key)
    df.rename(columns= {'United States                  not seasonally adjusted':'United States'}, inplace= True)
    df["United States"]= (df["United States"]-df["United States"][0]) * 100.0 / df["United States"][0]
    return df
```
