```python
df['100_mov_avg'] = df['Adj Close'].rolling(window = 100, min_periods = 0).mean()
```

```python
fig = plt.figure(figsize= (12,6))

ax1 = plt.subplot2grid((6,1), (0,0), rowspan = 5, colspan =1)
ax2 = plt.subplot2grid((6,1), (5,0), rowspan = 1, colspan =1, sharex = ax1)

ax1.plot(df.index, df['Adj Close'])
ax1.plot(df.index, df['100_mov_avg'])
ax2.bar(df.index, df['Volume'])
```

```python
df.drop('100_mov_avg', axis = 1, inplace = True)
df.head()
```

```python
df_ohlc = df['Adj Close'].resample('10D').ohlc()
```

```python
df_vol = df['Volume'].resample('10D').sum()
```

```python
df_ohlc.head()
```

```python
from matplotlib.finance import candlestick_ohlc
import matplotlib.dates as mdates
```

```python
df_ohlc.reset_index(inplace=True)
df_ohlc.head()
```

```python
df_ohlc['Date'] = df_ohlc['Date'].map(mdates.date2num)
df_ohlc.head()
```

```python
fig = plt.figure(figsize= (12,6))

ax1 = plt.subplot2grid((6,1), (0,0), rowspan = 5, colspan =1)
ax2 = plt.subplot2grid((6,1), (5,0), rowspan = 1, colspan =1, sharex = ax1)

ax1.xaxis_date()
candlestick_ohlc(ax1, df_ohlc.values, width = 2,  colorup='g')
ax2.fill_between(df_vol.index.map(mdates.date2num), df_vol.values, 0)
```
